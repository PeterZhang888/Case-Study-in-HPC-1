### Function: Poisson2D(n)
The function Poisson2D(n) constructs the sparse matrix $A$ for the two-dimensional Poisson problem
$$
-\Delta u = f
$$
on the unit square $(0,1)^2$, with homogeneous Dirichlet boundary conditions

$$
u(x,y)=0 \quad \text{for } x=0,\ x=1,\ y=0,\ \text{or } y=1.
$$
The domain is discretised using second-order finite differences on a uniform $n \times n$ interior grid. Since there are $n$ interior points in the $x$-direction and $n$ interior points in the $y$-direction, the total number of unknowns is
$$
n^2.
$$

Therefore, the matrix $A$ has size
$$
n^2 \times n^2.
$$

The matrix corresponds to the standard five-point finite difference stencil for the operator $-\Delta$.
#### Input
n : Number of interior grid points in each spatial direction.

#### Output
A: Sparse matrix of size $n^2 \times n^2$ representing the five-point finite difference discretisation of the 2D Poisson problem.

```python
def Poisson2D(n):
   
    h = 1.0 / (n + 1)
    # 1D second-difference matrix for -d^2/dx^2
    main_diag = 2.0 * np.ones(n)
    off_diag = -1.0 * np.ones(n - 1)
    T = sp.diags(
        diagonals=[off_diag, main_diag, off_diag],
        offsets=[-1, 0, 1],
        shape=(n, n),
        format="csr"
    )
    
    # Identity matrix
    I = sp.eye(n, format="csr")
    
    # 2D Poisson matrix:
    # A = (I ⊗ T + T ⊗ I) / h^2
    A = (sp.kron(I, T) + sp.kron(T, I)) / h**2
    
    return A.tocsr()
```
